```python
import cv2
import numpy as np

# Load YOLOv3 weights and configuration files
net = cv2.dnn.readNet("yolov3.weights", "yolov3.cfg")

# Load class names
classes = []
with open("coco.names", "r") as f:
    classes = [line.strip() for line in f.readlines()]

# Set input and output layers
layer_names = net.getLayerNames()
output_layer_names = [layer_names[i - 1] for i in net.getUnconnectedOutLayers()]

# Initialize camera
cap = cv2.VideoCapture(0)

# Loop through frames from the camera
while True:
    ret, frame = cap.read()
    
    if not ret:
        break
    
    # Resize the frame to a smaller size
    frame = cv2.resize(frame, (640, 480))
    
    # Get height, width, and channels of the frame
    height, width, channels = frame.shape
    
    # Prepare frame for inference
    blob = cv2.dnn.blobFromImage(frame, 1/255.0, (416, 416), swapRB=True, crop=False)

    # Pass frame through the network
    net.setInput(blob)
    outs = net.forward(output_layer_names)

    # Process detections
    class_ids = []
    confidences = []
    boxes = []
    for out in outs:
        for detection in out:
            scores = detection[5:]
            class_id = np.argmax(scores)
            confidence = scores[class_id]
            if confidence > 0.5:
                # Object detected
                center_x = int(detection[0] * width)
                center_y = int(detection[1] * height)
                w = int(detection[2] * width)
                h = int(detection[3] * height)
                x = int(center_x - w / 2)
                y = int(center_y - h / 2)
                boxes.append([x, y, w, h])
                confidences.append(float(confidence))
                class_ids.append(class_id)

    # Apply non-maximum suppression to remove overlapping boxes
    indexes = cv2.dnn.NMSBoxes(boxes, confidences, 0.5, 0.4)

    # Draw boxes and labels on the frame
    font = cv2.FONT_HERSHEY_PLAIN
    colors = np.random.uniform(0, 255, size=(len(classes), 3))
    if len(indexes) > 0:
        for i in indexes.flatten():
            x, y, w, h = boxes[i]
            label = classes[class_ids[i]]
            confidence = confidences[i]
            color = colors[class_ids[i]]
            cv2.rectangle(frame, (x, y), (x+w, y+h), color, 2)
            text = f"{label} {confidence:.2f}"
            cv2.putText(frame, text, (x, y-5), font, 2, color, 2)

    # Display the frame
    cv2.imshow("Object Detection", frame)

    # Exit if 'q' key is pressed
    if cv2.waitKey(1) == ord('0'):
        break

# Release resources
cap.release()
cv2.destroyAllWindows()
```

```python
""" The code performs real-time object detection using the YOLOv3 model on a video stream from the camera.

 Here's a breakdown of the code:

 1. Imports the necessary libraries, including OpenCV (`cv2`) and NumPy (`np`).

 2. Loads the pre-trained YOLOv3 weights and configuration files using `cv2.dnn.readNet()`. The weights file (`yolov3.weights`) 
 contains the learned parameters of the model, while the configuration file (`yolov3.cfg`) specifies the architecture of the 
 network.

 3. Loads the class names from the file `coco.names` into a list.

 4. Sets up the input and output layers of the network by obtaining the layer names from the network using 
 `net.getLayerNames()` and selecting the unconnected output layers using `net.getUnconnectedOutLayers()`.

 5. Initializes the camera by creating a `VideoCapture` object (`cap`) and setting it to capture frames from the default camera 
 (index 0).

 6. Enters a loop that continuously reads frames from the camera using `cap.read()`.

 7. Resizes each frame to a smaller size (640x480) to speed up processing.

 8. Prepares the frame for inference by converting it to a blob using `cv2.dnn.blobFromImage()`. The blob is a standardized 
 input format that the network expects, and it involves resizing, scaling, and normalizing the frame.

 9. Passes the prepared frame through the network using `net.setInput()` and `net.forward()` to obtain the output predictions 
 (`outs`).

 10. Processes the detections by iterating over the output predictions and filtering out the detections below a confidence 
 threshold. The bounding box coordinates, confidence scores, and class IDs of the detected objects are stored in lists.

 11. Applies non-maximum suppression using `cv2.dnn.NMSBoxes()` to remove overlapping bounding boxes and keep only the most 
 confident and non-overlapping detections. The indexes of the selected detections are stored in the `indexes` variable.

 12. Draws bounding boxes and labels on the frame for the selected detections using a random color for each class. The class 
 name and confidence score are displayed with the bounding boxes.

 13. Displays the annotated frame in a window using `cv2.imshow()`.

 14. Checks for the '0' key press using `cv2.waitKey(1)`. If the key is pressed, the loop is exited, and the program terminates.

 15. Releases the resources by releasing the camera (`cap.release()`) and destroying all OpenCV windows 
 (`cv2.destroyAllWindows()`).

 Overall, this code continuously captures frames from the camera, performs object detection on each frame using the YOLOv3 
 model, and displays the annotated frames in real-time. It allows you to visualize the detected objects and their labels in 
 a video stream from the camera."""
```
